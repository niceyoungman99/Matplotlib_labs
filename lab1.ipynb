**Четвертый пункт**   
Для параметров (B=0.7 (M2, M1): (0.06, 0.06))

```python
def calculate_3(B, M1, M2, x, y, z):
    x_new = y
    y_new = z
    z_new = M1 + B * x + M2 * y - z ** 2.0
    return x_new, y_new, z_new


B = 0.7
M1 = 0.06
M2 = 0.06
x, y, z = 0.1, 0.1, 0.1
x_mass3 = []
y_mass3 = []
z_mass3 = []
for i in range(1000):
    x, y, z = calculate_3(B, M1, M2, x, y, z)
    x_mass3.append(x)
    y_mass3.append(y)
    z_mass3.append(z)
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')


ax.scatter(x_mass3, y_mass3, z_mass3, c='r', marker='o')


ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```

**Пятый пункт**    
Постройка бифуркационного дерева 

```python
import numpy as np
import matplotlib.pyplot as plt

def logistic_map(r, x, iterations=10000):
    values = []
    for i in range(iterations):
        x = r * x * (1 - x)
        values.append(x)
    return values[-1000:]

x = 0.5
r_values = np.linspace(2.4, 4.0, 1000)
x_values = []

plt.figure(figsize=(8, 8))
for r in r_values:
    x_values = logistic_map(r, x)
    plt.plot([r] * 1000, x_values, 'b.', markersize=1)

plt.xlabel("r")
plt.ylabel("x")
plt.title("Бифуркационное дерево для логистического отображения")
plt.show()
```
